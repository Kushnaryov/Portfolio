## Creating new features

```python
for dataset in combine:
    dataset['Title'] = dataset.Name.str.extract('([A-Za-z]+)\.', expand = False)
```

```python
pd.crosstab(train_df['Title'], train_df['Sex'])
```

```python
#replacing rare titles
for dataset in combine:
    dataset['Title'] = dataset['Title'].replace(['Capt', 
                                                     'Col', 
                                                     'Countess', 
                                                     'Don', 
                                                     'Jonkheer',
                                                     'Major',
                                                     'Rev',
                                                     'Dr',
                                                     'Sir',
                                                     'Lady'
                                                    ], 'Rare')
    dataset['Title'] = dataset['Title'].replace(['Mlle', 'Ms'], 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')
```

```python
pd.crosstab(train_df['Title'], train_df['Sex'])
```

```python
train_df[['Title', 'Survived']].groupby('Title', as_index = False).mean().sort_values('Survived', ascending = False)
```

```python
title_mapping = {'Mrs':1, 'Miss':2, 'Master': 3, 'Rare': 4, 'Mr': 5}
sex_mapping = {'male': 1, 'female': 2}
for dataset in combine:
    dataset['Sex'] = dataset['Sex'].map(sex_mapping)
    dataset['Title'] = dataset['Title'].map(title_mapping)
    dataset['Title'] = dataset['Title'].fillna(0)
    dataset['Title'] = dataset['Title'].astype(int)
combine = [train_df, test_df]
```

```python
train_df = train_df.drop(['Name', 'PassengerId'], axis = 1)
test_df = test_df.drop(['Name'], axis = 1)
combine = [train_df, test_df]
```

```python
train_df.head()
```

### Guessing Age

```python
guess_ages = np.zeros((2,3))
guess_ages
```

```python
for i in range(0,2):
    for j in range(0,3):
        guess_df = train_df[(train_df['Sex'] == i+1) & (train_df['Pclass'] == j+1)]['Age'].dropna()
        age_guess = guess_df.mean()
        guess_ages[i,j] = int(age_guess)
```

```python
guess_ages
```

```python
for i in range(2):
    for j in range(3):
        train_df.loc[(train_df.Age.isnull()) & (train_df.Sex == i+1) & (train_df.Pclass == j+1),'Age'] = guess_ages[i,j]
        test_df.loc[(test_df.Age.isnull()) & (test_df.Sex == i+1) & (test_df.Pclass == j+1),'Age'] = guess_ages[i,j]
```

```python
train_df['Age'] = train_df['Age'].astype(int)
test_df['Age'] = test_df['Age'].astype(int)
combine = [train_df, test_df]
```

```python
train_df['AgeBand'] = pd.cut(train_df['Age'],5)
train_df[['AgeBand','Survived']].groupby('AgeBand', as_index = False).mean().sort_values(by = 'AgeBand')
```

```python
for dataset in combine:
    dataset.loc[dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32) , 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48) , 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64) , 'Age'] = 3
    dataset.loc[dataset['Age'] > 64 , 'Age'] = 4
train_df.head()
    
```

```python
train_df = train_df.drop(['AgeBand'], axis = 1)
combine = [train_df, test_df]
train_df.head()
```

```python
for dataset in combine:
    dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch']
train_df[['FamilySize', 'Survived']].groupby('FamilySize').mean().sort_values(by='Survived', ascending = False)
```

```python
for dataset in combine:
    dataset['IsAlone'] = 0
    dataset.loc[dataset['FamilySize'] == 0, 'IsAlone'] = 1

train_df[['IsAlone', 'Survived']].groupby('IsAlone').mean().sort_values(by='Survived', ascending = False)
combine = [train_df, test_df]
```

```python
try:
    test_df =test_df.drop(['Parch', 'SibSp', 'FamilySize'], axis = 1)
except:
    pass
try:
    train_df =train_df.drop(['Parch', 'SibSp', 'FamilySize'], axis = 1)
except:
    pass
combine = [train_df, test_df]
```

```python
train_df.head()
```

```python
freq_port = train_df.Embarked.dropna().mode()[0]
freq_port
```

```python
train_df['Embarked']=train_df['Embarked'].fillna(freq_port)
test_df['Embarked']=test_df['Embarked'].fillna(freq_port)
train_df[['Embarked', 'Survived']].groupby('Embarked').mean().sort_values(by='Survived', ascending = False)
combine = [train_df, test_df]
```

```python
train_df['Embarked'] = train_df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2}).astype(int)
test_df['Embarked'] = test_df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2}).astype(int)

train_df.head()
combine = [train_df, test_df]
```

```python
test_df['Fare'].fillna(test_df['Fare'].dropna().median(), inplace = True)
test_df.head()
combine = [train_df, test_df]
```

```python
train_df['FareBand'] = pd.qcut(train_df['Fare'], 4)
train_df[['FareBand', 'Survived']].groupby('FareBand').mean().sort_values(by='Survived', ascending = True)
combine = [train_df, test_df]
```

```python
for dataset in combine:
    dataset.loc[dataset['Fare'] <= 7.91, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454) , 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31.0) , 'Fare'] = 2
    dataset.loc[dataset['Fare'] > 31.0 , 'Fare'] = 3
    dataset['Fare'] = dataset['Fare'].astype(int)
train_df = train_df.drop('FareBand', axis = 1)
combine = [train_df, test_df]
```
